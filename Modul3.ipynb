```python
#1.Masukkan iris.csv ke dataframe
import pandas as pd
df = pd.read_csv('iris.csv') #membaca file iris.csv

#print(df.to_string())
df
```

```python
#2.Normalisasikan setiap attribut iris ke dataframe menggunakan normalisasi min max sehingga setiap nilai attribut memiliki range 0 – 1. 
#Vbaru = (Vlama - Vmin) / (Vmax - Vmin). Vbaru = nilai data di suatu sel setelah normalisasi, Vlama = nilai data asli sebelum normalisasi, 
#Vmax = nilai max dari kolom data tersebut, Vmin = nilai min dari kolom data tersebut 

spl_max = df['sepal.length'].max()
spl_min = df['sepal.length'].min()
ptl_max = df['petal.length'].max()
ptl_min = df['petal.length'].min()
spw_max = df['sepal.width'].max()
spw_min = df['sepal.width'].min()
ptw_max = df['petal.width'].max()
ptw_min = df['petal.width'].min()

for i in df.index :
  df.loc[i, 'sepal.length'] = (df.loc[i, 'sepal.length']-spl_min)/(spl_max-spl_min)
  df.loc[i, 'petal.length'] = (df.loc[i, 'petal.length']-ptl_min)/(ptl_max-ptl_min)
  df.loc[i, 'sepal.width'] = (df.loc[i, 'sepal.width']-spw_min)/(spw_max-spw_min)
  df.loc[i, 'petal.width'] = (df.loc[i, 'petal.width']-ptw_min)/(ptw_max-ptw_min)

print(df.to_string())
```

```python
#3.Cari centroid dari data yang telah dinormalisasi. Centroid dihitung dari rata2 seluruh data suatu kolom
print ('====SETOSA====')
spl_setosa = df[df.variety == 'Setosa']['sepal.length'].mean() #rata2
print (spl_setosa)
spw_setosa = df[df.variety == 'Setosa']['sepal.width'].mean()
print (spw_setosa)
ptl_setosa = df[df.variety == 'Setosa']['petal.length'].mean()
print (ptl_setosa)
ptw_setosa = df[df.variety == 'Setosa']['petal.width'].mean()
print (ptw_setosa)

print ('====VERSICOLOR====')
spl_versicolor = df[df.variety == 'Versicolor']['sepal.length'].mean()
print (spl_versicolor)
spw_versicolor = df[df.variety == 'Versicolor']['sepal.width'].mean()
print (spw_versicolor)
ptl_versicolor = df[df.variety == 'Versicolor']['petal.length'].mean()
print (ptl_versicolor)
ptw_versicolor = df[df.variety == 'Versicolor']['petal.width'].mean()
print (ptw_versicolor)

print ('====VIRGINICA====')
spl_virginica = df[df.variety == 'Virginica']['sepal.length'].mean()
print (spl_virginica)
spw_virginica = df[df.variety == 'Virginica']['sepal.width'].mean()
print (spw_virginica)
ptl_virginica = df[df.variety == 'Virginica']['petal.length'].mean()
print (ptl_virginica)
ptw_virginica = df[df.variety == 'Virginica']['petal.width'].mean()
print (ptw_virginica)
```

```python
#4.Masukkan iristesting.csv ke dataframe. Normalisasikan iristesting.csv menggunakan vmin dan vmax pada data 1

df_test = pd.read_csv('iristesting.csv') #membaca file iristesting.csv
df_test

def normalisasi(data1):
  return (df_test[data1] - df_test[data1].min())/(df_test[data1].max() - df_test[data1].min())

df_test['sepal.length'] = normalisasi('sepal.length')
df_test['sepal.width'] = normalisasi('sepal.width')
df_test['petal.length'] = normalisasi('petal.length')
df_test['petal.width'] = normalisasi('petal.width')
df_test
```

```python
#5.Prediksilah kelas data di iristesting.csv menggunakan kriteria kedekatan distance antara setiap data di iristesting.csv 
#ke centroid masing2 class di iris. Gunakan euclidean distance

import pandas as pd
import numpy as np
df = pd.read_csv('iristesting.csv')
atr_iristesting = ['sepal.length', 'sepal.width', 'petal.length', 'petal.width']

def normalisasi(atr_iristesting):
  return (df[atr_iristesting] - df[atr_iristesting].min()) / (df[atr_iristesting].max() - df[atr_iristesting].min())

df["sepal.length"] = normalisasi("sepal.length")
df["sepal.width"] = normalisasi("sepal.width")
df["petal.length"] = normalisasi("petal.length")
df["petal.width"] = normalisasi("petal.width") 

df_last = pd.read_csv('iristesting.csv')

def normalisasi(atr_iristesting):
  return (df_test[atr_iristesting] - df_last[atr_iristesting].min()) / (df_last[atr_iristesting].max() - df_last[atr_iristesting].min())

df_test["sepal.length"] = normalisasi("sepal.length")
df_test["sepal.width"] = normalisasi("sepal.width")
df_test["petal.length"] = normalisasi("petal.length")
df_test["petal.width"] = normalisasi("petal.width")

def euclidian(data_ke, data_frame):
  x = ['setosa', 'versicolor', 'virginica']
  y = []
  hasil = None
  for i in x:
    kelas = df[df['variety'] == i]
    centroid = kelas.mean()
    test = data_frame.iloc[data_ke, 0:4]
    a = ((centroid[0]-test[0])**2 + (centroid[1]-test[1])**2 + (centroid[2]-test[2])**2 + (centroid[3]-test[3])**2)*1/2 #menghitung euclidian
    y.append(a)
  if (min(y)== y[0]):
    hasil = ('Setosa')
    return hasil
  elif (min(y)== y[1]):
    hasil = ('Versicolor')
    return hasil
  else :
    hasil = ('Virginica')
    return hasil

temp = [] #menampung hasil looping
for j in df_test.index:
  temp.append(euclidian (j,df_test))
print(df_test['variety'] == temp)
print(df_test['variety'])
```

```python
#6.Hitunglah berapa persen klasifikasi yang benar (akurasi) saat memprediksi class 
Vtrue = 0 #variabel yg true
Vfalse = 0 #variabel yg false
y = df_test['variety']== temp #ada di no 5

for x in df_test.index:
  if y[x] == True:
    Vtrue +=1
  else:
    Vfalse +=1
persen = (Vtrue/(Vfalse+Vtrue))*100
print (persen,"%")
```
