```python
Jawab
1. Dataset adalah kumpulan dari objek-objek data atau yang sering disebut record, point, vector, 
    pattern, event, observation, case, atau bahkan data.Terdapat bermacam-macam cara untuk merepresentasikan dataset   
    contohnya yaitu atribut yang digunakan untuk menggambarkan jenis objek (bisa berupa 
    kulitatif maupun kuantitatif). Data Training adalah Data training digunakan untuk melatih algoritma, 
    sedangkan Data Testing dipakai untuk mengetahui
    performa algoritma yang sudah dilatih sebelumnya ketika menemukan data baru yang belum pernah dilihat sebelumnya.
```

```python
2. a. Agregasi : penggabungan beberapa atribut data menjadi sebuah atribut
   b. Sampling : pengambilan sebagian data yang mewakili keseluruhan populasi
   c. Dimensionality reduction : pengurangan jumlah atribut data. Hanya menggunakan atribut atribut yang penting
   d. Feature Subset Selection: cara lain untuk mengurangi dimensi dari data 
   e. Feature creation : membuat atribut baru yang mungkin penting untuk analisa data
   f. Discretization and Binarization : Discreatization adalah transformasi atribut kontinyu kedalam bentuk atribut 
      katagorikal. Sedangkan, Binarization adalah transformasi atribut kontinyu maupun diskrit kedalam bentuk atribut biner
   g. Atribute Transformation : sebuah fungsi yang memetakan seluruh rangkaian nilai dari atribut tertentu ke satu set nilai 
      pengganti baru sehingga setiap nilai lama dapat diidentifikasi dengan salah satu nilai baru
```
